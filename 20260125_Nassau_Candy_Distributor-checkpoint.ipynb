# **1. Introducton**
Nassau Candy is one of the largest U.S. wholesale manufacturers and distributors of specialty and private-label confections, supplying national retailers and independent stores through an extensive portfolio of 20,000+ SKUs. With vertically integrated manufacturing, nut roasting, private-label customization, and nationwide distribution, the company operates at significant scale and complexity. However, in such environments, **sales volume alone is an incomplete indicator of business success.**

High-selling products may generate weak margins, consume excessive operational cost, or mask underlying profitability issues. This analysis focuses on moving beyond revenue metrics to establish product and division level profitability clarity, enabling data-driven decision making.

Key objectives of this analysis include:

- Identifying products that truly drive gross profit and margin

- Detecting high-revenue but low-margin products that weaken overall performance

- Evaluating profitability differences across product divisions

- Highlighting margin risks to support pricing, sourcing, and portfolio optimization decisions

# **2. Problem Statement**

Despite strong sales performance across multiple product lines, Nassau Candy currently lacks clear visibility into the **true profitability drivers** of its portfolio. Decision making around pricing, promotions, and product mix is therefore constrained by limited margin-level insight.

Specifically, the organization does not have a consolidated view of:
- Which product lines deliver the highest gross margins  
- Whether high-revenue products generate proportional profit  
- How profitability and margin efficiency vary across divisions  
- Which products pose margin risk due to high cost or low efficiency  

Addressing these gaps is critical to improving financial performance and ensuring that growth is driven by **profitable, sustainable products** rather than volume alone.

```python
# This function flags extreme values without modifying the dataset
def flag_outliers_iqr(
    df,
    exclude_cols=None,
    multiplier=1.5,
    suffix='_Outlier'
):
    # Default exclusion list
    if exclude_cols is None:
        exclude_cols = []

    # Automatically select numeric columns except excluded ones
    numeric_cols = (
        df
        .select_dtypes(include=['int', 'float'])
        .columns
        .difference(exclude_cols)
    )

    # Compute quartiles
    Q1 = df[numeric_cols].quantile(0.25)
    Q3 = df[numeric_cols].quantile(0.75)
    IQR = Q3 - Q1

    # Define bounds
    lower_bound = Q1 - multiplier * IQR
    upper_bound = Q3 + multiplier * IQR

    # Flag outliers (True = Outlier)
    outlier_flags = (df[numeric_cols] < lower_bound) | (df[numeric_cols] > upper_bound)

    # Append flags to original DataFrame
    return df.join(outlier_flags.add_suffix(suffix))
```

```python
# Using the newly created function to check number of outliers
df = flag_outliers_iqr(
    df,
    exclude_cols=['Row ID']  # add more IDs here if needed
)

# Quick sanity check
df.filter(like='_Outlier').sum()
```

```python
#Rows where any core metric is an outlier
outlier_mask = (
    df['Sales_Outlier'] |
    df['Units_Outlier'] |
    df['Gross Profit_Outlier'] |
    df['Cost_Outlier']
)

outlier_df = df.loc[outlier_mask, [
    'Product Name',
    'Sales',
    'Units',
    'Gross Profit',
    'Cost',
    'Sales_Outlier',
    'Units_Outlier',
    'Gross Profit_Outlier',
    'Cost_Outlier'
]]

outlier_df.sample(5)
```

```python
# Create a capped copy of the original dataframe
df_cap = df.copy()

# Columns to treat for outliers
outlier_cols = ['Sales', 'Units', 'Gross Profit', 'Cost']

# Apply capping
for col in outlier_cols:
    
    lower_limit = df[col].quantile(0.02)
    upper_limit = df[col].quantile(0.98)
    
    df_cap[col] = (
        df_cap[col]
        .astype(float)
        .clip(lower=lower_limit, upper=upper_limit)
    )
```

## **6.1 Gross Margin**
Gross Margin (%) = Gross Profit ÷ Sales
- **Gross Margin** measures profit efficiency relative to sales.

```python
df['Gross Margin'] = df['Gross Profit'] / df['Sales']
```

## **6.2 Profit per Unit**
Profit per Unit = Gross Profit ÷ Units
- **Profit per Unit** evaluates profitability at the unit level, independent of sales volume.

```python
df['Profit per Unit'] = df['Gross Profit'] / df['Units']
```

## **6.3 Cost Ratio**
Cost Ratio = Cost ÷ Sales
- **Cost Ratio** indicates how much of sales revenue is consumed by cost.

```python
df['Cost Ratio'] = df['Cost']/df['Sales']
```

## **6.4 Revenue per Unit**
Revenue per Unit = Sales ÷ Units
- Revenue per Unit reflects effective pricing per unit sold.

```python
df['Revenue Per Unit'] = df['Sales']/df['Units']
```

### **Product Segmentation**

```python
# Aggregate transactional data to product level to compute core metrices
product_metrics = df.groupby('Product Name').agg({
    'Sales': 'sum',
    'Units': 'sum',
    'Gross Profit': 'sum',# Define a percentile-based bucketing function to classify products into Low, Mid, High tiers
    'Cost': 'sum',
    'Gross Margin': 'mean'
}).reset_index()
```

```python
# Define a percentile-based bucketing function to classify products into Low, Mid, High tiers
def assign_percentile_level(series):
    return pd.qcut(
        series,
        q=[0, 0.25, 0.75, 1.0],
        labels=['Low', 'Mid', 'High']
    )

# Apply percentile-based segmentation to sales, profit, and margin metrics
product_metrics['Sales_Level'] = assign_percentile_level(product_metrics['Sales'])
product_metrics['Profit_Level'] = assign_percentile_level(product_metrics['Gross Profit'])
product_metrics['Margin_Level'] = assign_percentile_level(product_metrics['Gross Margin'])
```

```python
# Classify products into business-relevant segments based on combined sales, profit, and margin levels
def classify_product(row):

    if row['Profit_Level'] == 'High' and row['Margin_Level'] == 'High':
        return 'High-Profit / High-Margin'

    elif row['Sales_Level'] == 'High' and row['Margin_Level'] == 'Low':
        return 'High-Sales / Low-Margin'

    elif row['Sales_Level'] == 'Low' and row['Profit_Level'] == 'Low':
        return 'Low-Sales / Low-Profit'

    else:
        return 'Other'
```

```python
# Apply the product segmentation logic across all products
product_metrics['Product_Segment'] = product_metrics.apply(
    classify_product,
    axis=1
)
```

```python
# Summarize the number of products in each business segment
product_metrics['Product_Segment'].value_counts()
```

## **7.3 Division-Level Performance Analysis**

### **Average margin by division**

```python
df['Division'].value_counts()
```

```python
# Aggregate sales, cost, and profit metrics at the division level
division_metrics = df.groupby('Division').agg({
    'Sales': 'sum',
    'Gross Profit': 'sum',
    'Cost': 'sum'
}).reset_index()

# Calculate gross margin at the division level
division_metrics['Gross Margin'] = (
    division_metrics['Gross Profit'] / division_metrics['Sales']
)
```

```python
# Calculate revenue and profit contribution percentages
division_metrics['Revenue_Contribution_%'] = (
    division_metrics['Sales'] / division_metrics['Sales'].sum()
) * 100

division_metrics['Profit_Contribution_%'] = (
    division_metrics['Gross Profit'] / division_metrics['Gross Profit'].sum()
) * 100

# Compute imbalance between revenue and profit contribution
division_metrics['Profit_Revenue_Gap'] = (
    division_metrics['Profit_Contribution_%'] -
    division_metrics['Revenue_Contribution_%']
)
```

```python
# Plot average gross margin by division
plt.figure(figsize=(10,6))

sns.barplot(
    data=division_metrics.sort_values('Gross Margin', ascending=False),
    color='#4C72B0',
    x='Division',
    y='Gross Margin'
)

plt.title('Average Gross Margin by Division', fontsize=14, fontweight='bold')
plt.xlabel('Division')
plt.ylabel('Gross Margin (%)')
plt.grid(axis='y', linestyle=':', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# Prepare data for grouped bar chart
divisions = division_metrics['Division']
revenue_pct = division_metrics['Revenue_Contribution_%']
profit_pct = division_metrics['Profit_Contribution_%']

x = range(len(divisions))
bar_width = 0.35

# Create figure
plt.figure(figsize=(10,6))

# Plot revenue and profit contribution bars
plt.bar(
    x,
    revenue_pct,
    width=bar_width,
    label='Revenue Contribution (%)',
    color = '#4C72B0'
)

plt.bar(
    [i + bar_width for i in x],
    profit_pct,
    width=bar_width,
    label='Profit Contribution (%)',
    color = '#4CB06F'
)

# Annotate percentage values on bars
for i in range(len(divisions)):
    plt.text(
        i,
        revenue_pct.iloc[i] + 0.5,
        f"{revenue_pct.iloc[i]:.1f}%",
        ha='center',
        fontsize=10
    )
    plt.text(
        i + bar_width,
        profit_pct.iloc[i] + 0.5,
        f"{profit_pct.iloc[i]:.1f}%",
        ha='center',
        fontsize=10
    )

# Axis labels and title
plt.xticks(
    [i + bar_width / 2 for i in x],
    divisions,
    fontweight='bold'
)

plt.title('Revenue vs Profit Contribution by Division', fontsize=14, fontweight='bold')
plt.ylabel('Contribution Percentage (%)')
plt.legend()
plt.grid(axis='y', linestyle=':', alpha=0.6)

plt.tight_layout()
plt.show()
```

## **7.5 Cost Structure Diagnostics**

### **Cost Structure Diagnostics**

```python
# Initialize figure
plt.figure(figsize=(12, 6), dpi=300)

# Scatter plot: Cost vs Sales, colored by Margin Level
sns.scatterplot(
    data=product_metrics,
    x='Cost',
    y='Sales',
    hue='Margin_Level',        # Low / Mid / High
    palette={'Low':'#d62728', 'Mid':'#ff7f0e', 'High':'#1f77b4'},
    s=80,
    alpha=0.7
)

# Axis formatting
plt.xlabel('Product Cost ($)', fontsize=12, fontweight='bold')
plt.ylabel('Sales ($)', fontsize=12, fontweight='bold')
plt.title('Cost vs Sales: Identify Margin Pressure', fontsize=14, fontweight='bold')

# Professional grid (y-axis only)
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.legend(title='Margin Level', fontsize=10, title_fontsize=11)

plt.tight_layout()
plt.show()
```

```python
# Precompute portfolio-level thresholds for cost and sales
cost_median = product_metrics['Cost'].median()
sales_median = product_metrics['Sales'].median()

# Flag products based on cost, margin, and sales efficiency
def flag_cost_issues(row):
    if row['Gross Margin'] <= 0.20 and row['Cost'] > cost_median:
        return 'Repricing'
    elif row['Gross Margin'] <= 0.35 and row['Cost'] > cost_median:
        return 'Cost renegotiation'
    elif row['Gross Margin'] <= 0.20 and row['Sales'] < sales_median:
        return 'Discontinuation review'
    else:
        return 'No immediate action'

# Apply cost structure diagnostic classification
product_metrics['Cost_Action_Flag'] = product_metrics.apply(
    flag_cost_issues,
    axis=1
)

# Review distribution of cost action flags
product_metrics['Cost_Action_Flag'].value_counts()
```

```python
# Cost vs Sales diagnostic to identify pricing and cost inefficiencies
plt.figure(figsize=(12, 6), dpi=300)

sns.scatterplot(
    data=product_metrics,
    x='Cost',
    y='Sales',
    hue='Cost_Action_Flag',
    palette={
        'Repricing': '#d62728',
        'Cost renegotiation': '#ff7f0e',
        'Discontinuation review': '#9467bd',
        'No immediate action': '#1f77b4'
    },
    s=80,
    alpha=0.75
)

plt.xlabel('Total Cost ($)', fontsize=12, fontweight='bold')
plt.ylabel('Total Sales ($)', fontsize=12, fontweight='bold')
plt.title('Cost vs Sales Diagnostic: Pricing and Cost Efficiency', fontsize=14, fontweight='bold')

# Professional grid (y-axis only)
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.legend(title='Recommended Action', fontsize=10, title_fontsize=11)

plt.tight_layout()
plt.show()
```
